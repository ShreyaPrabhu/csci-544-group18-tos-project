```python
from transformers import AutoTokenizer, AutoModelForMaskedLM

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

model = AutoModelForMaskedLM.from_pretrained("bert-base-uncased")
```

```python
sample = 'where is Himalayas in the world map?'
encoding = tokenizer.encode(sample)
print(encoding)
print(tokenizer.convert_ids_to_tokens(encoding))
```

Notice "##s" this is a way to handle unknown words in Wordpiece.

```python
sample = 'where is Himalayass in the world map? [SEP] This is multiline'
encoding = tokenizer.encode(sample)
print(encoding)
print(tokenizer.convert_ids_to_tokens(encoding))
```

```python
q1 = 'Who was Tony Stark?'
c1 = 'Anthony Edward Stark known as Tony Stark is a fictional character in Avengers'
q2 = 'Who was Tony in Marvel'
c2 = 'Tony Stark is a fictional character in Marvel Avengers'
encoding = tokenizer([q1,q2], [c1,c2])
for key, value in encoding.items():
    print('1- {}: {}'.format(key, value))
```

```python
encoding = tokenizer([q1,q2], [c1,c2],padding=True)
for key, value in encoding.items():
    print('1- {}: {}'.format(key, value))
```

```python
import csv
```

```python
import pandas as pd
```

```python
data = []
with open(clauses_file, mode="r") as file:
  reader = csv.reader(file, delimiter=",")
  for row in reader:
    data.append(row[1])

df = pd.DataFrame(data[1:], columns=["sentences"])
df.head()
```

```python
encodings = []
for line in df["sentences"][:5]:
  encoding = tokenizer.encode(line)
  encodings.append(tokenizer.convert_ids_to_tokens(encoding))
```

```python
encodings = []
for line in df["sentences"]:
  encoding = tokenizer.encode(line[:512])
  encodings.append(encoding)
```
